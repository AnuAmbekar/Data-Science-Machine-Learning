```python
import pandas as pd
from sklearn.metrics import accuracy_score
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import sklearn.metrics as metrics
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import 1DecisionTreeRegressor
import pickle
import seaborn as sns
```

```python
df = pd.read_csv('Bengaluru_House_Data.csv')
```

```python
df=df.dropna()
```

```python
df['rooms'] = df['size'].apply(lambda x: x.split(' ')[0])
```

```python
df = df.drop(columns='size')
```

```python
df = df.reset_index(drop=True)
```

```python
df = df.drop(columns=['availability', 'balcony'])
```

```python
df1 = df.loc[df['total_sqft'].str.contains('-')]
df = df.loc[~df['total_sqft'].str.contains('-')]
```

```python
df1[['one', 'two']] = df1['total_sqft'].str.split('-', expand=True)
df1
```

```python
df1['one']=df1['one'].astype(float)
df1['two']=df1['two'].astype(float)

df1['total_sqft'] = (df1['one'] + df1['two']) / 2
```

```python
#df1 = df1.drop(columns=['total_sqft', 'one', 'two'])
```

```python
df1 = df1.drop(columns=['one', 'two'])
```

```python
df = pd.concat([df, df1])
```

```python
df['total_sqft'] = df['total_sqft'].astype(str)
df=df.loc[~df['total_sqft'].str.contains('Meter')]
df=df.loc[~df['total_sqft'].str.contains('Yards')]
df=df.loc[~df['total_sqft'].str.contains('Acre')]
```

```python
df['total_sqft']=df['total_sqft'].astype(float)
```

```python
df['area_sqft'] = df['price']*100000/df['total_sqft']
df
```

```python
df['area_sqft'].quantile([0.25, 0.5, 0.75])


df = df.loc[(df['area_sqft'] >= 4279.131007) & (df['area_sqft'] <= 6696.629213)]
df
```

```python
df = df.drop(columns='area_sqft')
```

```python
dummies = pd.get_dummies(df['location'])
dummies

le = LabelEncoder()

df['area_type'] = le.fit_transform(df['area_type'])
df['bath']=df['bath'].astype(int)
df['rooms'] = df['rooms'].astype(int)

df = pd.get_dummies(df, columns = ['location'])

x = df['price']
y = df.drop(columns=['price', 'society'])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)

x_train = (np.array(x_train)).reshape(-1, 1)
x_test = (np.array(x_test)).reshape(-1, 1)

lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```

```python
from sklearn import linear_model
lasso = linear_model.Lasso(alpha=0.1)
lasso.fit(x_train, y_train)
lasso.score(x_test, y_test)
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)
ridge.score(x_test, y_test)
```
