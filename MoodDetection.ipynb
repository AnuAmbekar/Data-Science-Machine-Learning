```python
import cv2
import mediapipe as mp
import numpy as np
import csv
import time
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import pickle
```

```python
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils
```

```python
video = cv2.VideoCapture(0)

start = time.time()

with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    
    while time.time() - start < 10:
        
        ret, frame = video.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        results = holistic.process(image)
        
        original_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        mp_drawing.draw_landmarks(original_image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION,
                                 mp_drawing.DrawingSpec((0, 255, 0), 2, 2), mp_drawing.DrawingSpec((0, 255, 0), 2, 2))
        
        mp_drawing.draw_landmarks(original_image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                                  mp_drawing.DrawingSpec((255, 255, 0), 2, 2), mp_drawing.DrawingSpec((255, 255, 0), 2, 2))
           
        mp_drawing.draw_landmarks(original_image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                                   mp_drawing.DrawingSpec((0, 255, 255), 2, 2), mp_drawing.DrawingSpec((0, 255, 255), 2, 2))
    
        mp_drawing.draw_landmarks(original_image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                                   mp_drawing.DrawingSpec((0, 255, 0), 2, 2), mp_drawing.DrawingSpec((0, 255, 0), 2, 2))
        
        cv2.imshow('image', original_image)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    video.release()
    cv2.destroyAllWindows()
```

```python
num_coords = len(results.pose_landmarks.landmark) + len(results.face_landmarks.landmark)
```

```python
landmarks = ['class']

for i in range(1, num_coords+1):
    landmarks += [f'x{i}', f'y{i}', f'z{i}', f'v{i}']
    
with open('coords.csv', 'w') as file:
    writer = csv.writer(file)
    
    writer.writerow(landmarks)
```

```python
df = pd.read_csv('coords.csv').dropna(axis=1)

X = df.drop(columns='class')
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
clf.score(X_test, y_test)

filename = 'model.sav'
pickle.dump(clf, open(filename, 'wb'))
```

```python
model = pickle.load(open(filename, 'rb'))
```

```python
video = cv2.VideoCapture(0)

start = time.time()

with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    
    while True:
        
        ret, frame = video.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        results = holistic.process(image)
        
        original_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        mp_drawing.draw_landmarks(original_image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION,
                                 mp_drawing.DrawingSpec((0, 255, 0), 2, 2), mp_drawing.DrawingSpec((0, 255, 0), 2, 2))
        
        mp_drawing.draw_landmarks(original_image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                                  mp_drawing.DrawingSpec((255, 255, 0), 2, 2), mp_drawing.DrawingSpec((255, 255, 0), 2, 2))
           
        mp_drawing.draw_landmarks(original_image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                                   mp_drawing.DrawingSpec((0, 255, 255), 2, 2), mp_drawing.DrawingSpec((0, 255, 255), 2, 2))
    
        mp_drawing.draw_landmarks(original_image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                                   mp_drawing.DrawingSpec((0, 255, 0), 2, 2), mp_drawing.DrawingSpec((0, 255, 0), 2, 2))
        
        pose = np.array(([(i.x, i.y, i.z, i.visibility) for i in results.pose_landmarks.landmark])).flatten()
        face = np.array(([(i.x, i.y, i.z) for i in results.face_landmarks.landmark])).flatten()
        rows = list(np.concatenate([pose, face]))
        prediction = model.predict(np.array(rows).reshape(1, -1))
        
        x=results.pose_landmarks.landmark[mp_holistic.PoseLandmark.RIGHT_EAR].x
        y=results.pose_landmarks.landmark[mp_holistic.PoseLandmark.RIGHT_EAR].y
        array = tuple(np.multiply(np.array([x,y]), [640, 480]).astype(int))
        
        cv2.putText(original_image, str(prediction), (array[0], array[1]), cv2.FONT_HERSHEY_SIMPLEX,  
                   1, (255, 255, 255) , 2, cv2.LINE_AA) 
        
        cv2.imshow('image', original_image)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    video.release()
    cv2.destroyAllWindows()
```
