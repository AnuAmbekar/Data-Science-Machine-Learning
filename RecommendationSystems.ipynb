```python
def add_product(url):
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

    page = requests.get(url, headers=headers)
    
    if not page.status_code:
        print("There was an error with the status code.")
    
    soup = bs(page.content)
    
    title = soup.find_all("span", class_="a-size-extra-large celwidget")
    author = soup.find_all("a", class_="a-size-base a-link-normal a-text-normal")
    
    data_dict = {"title": title, "author": author}
    
    df = pd.DataFrame(data_dict)
    
    df = df.astype(str)
    
    df["title"] = ((df["title"].apply(lambda x: x.split(">")[1])).apply(lambda x: x.split("<")[0])).apply(lambda x: x.strip())
    df["author"] = (df["author"].apply(lambda x: x.split(">")[1])).apply(lambda x: x.split("<")[0])
    
    return df
```

### Using nearest neighbours to recommend most similar books.


```python
vectorizer = CountVectorizer()

vector = vectorizer.fit_transform(df["title"])

matrix = vector.todense()

idx = pd.Series(df['title'], df.index)

book = 'Atomic Habits: the life-changing million-copy #1 bestseller'

index = df[df['title'] == book].index[0]

index = matrix[index]

nn = NearestNeighbors(n_neighbors=6, algorithm='ball_tree').fit(matrix)

distances, indices = nn.kneighbors(matrix)

nearest_distances = distances[0][1:]

nearest_indices = indices[0][1:]

df.iloc[nearest_indices]
```
