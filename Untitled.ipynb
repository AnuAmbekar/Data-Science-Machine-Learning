```python
import os
import re
import PIL
from PIL import Image
import pandas as pd
import glob
import shutil
import numpy as np
from numpy import asarray
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
initial_dir = os.getcwd()
working_dir = os.path.join(initial_dir, "train")
```

```python
fruit_data = []
filenames = []
folder_names = []
for folder, _, filename in os.walk(working_dir):
    folder_name = os.path.basename(folder)
    
    dictionary = {'fruit_name': folder_name, 'filename': filename}
    fruit_data.append(dictionary)
    folder_names.append(folder_name)
```

```python
len(folder_names)
```

```python
fruit_dfs = []


for i in range(len(folder_names)):
    
    fruits = fruit_data[i]['fruit_name']
    filename = fruit_data[i]['filename']
    
    data = {'fruit_name': fruits, 'filename': filename}
    data = pd.DataFrame(data)
    
    
    fruit_dfs.append(data)
```

```python
df = pd.concat(fruit_dfs, ignore_index=True)
```

```python
df
```

```python
new_working_dir = os.path.join(working_dir, "all_images")
```

```python
df = df.loc[~df['fruit_name'].str.contains('all_images')]
```

```python
df
```

```python
pixelated_array = []
fruit_names = []
for index, row in df.iterrows():
    filename = row['filename']
    filename = os.path.join(new_working_dir, filename)
    image = Image.open(filename)
    image.resize((256, 256))
    pixels = np.asarray(image).astype('float32')
    pixels /= 255.0
    pixelated_array.append(pixels)
```

```python
pixelated_array
```

```python
pixelated_array = np.array(pixelated_array)
```

```python
pixelated_array
```

```python
le = LabelEncoder()

df['fruit_name'] = le.fit_transform(df['fruit_name'])
```

```python
labels = df['fruit_name'].to_numpy()
len(labels)
len(pixelated_array)
```

```python
for s in pixelated_array:
    print(np.shape(s))
```

```python
X_train, X_test, y_train, y_test = train_test_split(pixelated_array, labels, test_size=0.2)
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split

model = keras.Sequential([
    keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(100, 100, 3)),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(64, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(128, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(33, activation='softmax')  # Adjust the number of output classes as needed
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',  # Use categorical cross-entropy for multiclass classification
              metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

# Evaluate the model on test data
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'Test accuracy: {test_acc}')
```
