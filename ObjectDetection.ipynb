```python
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
mp_drawing = mp.solutions.drawing_utils
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = 'efficientdet_lite2_uint8.tflite'

baseOptions = mp.tasks.BaseOptions(model)
objectDetectorOptions = mp.tasks.vision.ObjectDetectorOptions(base_options=baseOptions, score_threshold=0.5)
objectDetector = mp.tasks.vision.ObjectDetector.create_from_options(objectDetectorOptions)

mp_image = mp.Image.create_from_file('images.jpeg')
detections = objectDetector.detect(image=mp_image)
image = mp_image.numpy_view()
```

```python
for result in detections.detections:
    #Get the bounding box first.
    bounding_box = result.bounding_box
    
    #Create the rectangle.
    start_point = bounding_box.origin_x, bounding_box.origin_y
    end_point = bounding_box.origin_x + bounding_box.width, bounding_box.origin_y + bounding_box.height
    cv2.rectangle(image, start_point, end_point, (255, 255, 0), 1)
    print(result.categories[0].category_name)
    #Add the text
    cv2.putText(image, f'{result.categories[0].category_name}', (bounding_box.origin_x, bounding_box.origin_y), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1, cv2.LINE_AA) 
    result.categories??
plt.imshow(image)
```
