```python
import os
import pandas as pd
import io
import xml.etree.ElementTree as ET
from PIL import Image
from numpy import asarray  # Import the asarray function from NumPy
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.preprocessing import OneHotEncoder
import tensorflow as tf
```

```python
cwd = os.getcwd()
annotations = 'C://Users/anurag/Documents/Python/Sentiment Analysis/Mask Detection/annotations'

xml_files = []
for xml_file in os.listdir(annotations):
    xml_files.append(xml_file)
```

```python
filename = []
mask_status = []
xmin = []
ymin = []
xmax = []
ymax = []
```

```python
names = []

for file in xml_files:
    tree = ET.parse('C://Users/anurag/Documents/Python/Sentiment Analysis/Mask Detection/annotations/{}'.format(file))
    root = tree.getroot()
    wearing_mask = (root.find('object')).find('name').text
    mask_status.append(wearing_mask)
    
    
    for r in root.findall('object'):
        for b in r.findall('bndbox'):
            xmins = b.find('xmin').text
            xmin.append(xmins)
            ymins = b.find('ymin').text
            ymin.append(ymins)
            xmaxs = b.find('xmax').text
            xmax.append(xmaxs)
            ymaxs = b.find('ymax').text
            ymax.append(ymaxs)
    
    for obj in root.findall('object'):
        name = obj.find('name').text
        filenames = root.find('filename').text
        names.append(name)
        filename.append(filenames)
```

```python
print(len(filename))
print(len(xmin))
print(len(ymin))
print(len(xmax))
print(len(ymax))
print(len(names))
```

```python
data = {'mask_status': names, 'filename': filename, 'xmin': xmin, 'ymin': ymin, 'xmax': xmax, 'ymax': ymax}

df = pd.DataFrame(data)

df.head()
```

```python
df[['mask_status', 'filename']]=df[['mask_status', 'filename']].astype(str)
df
```

```python
images = 'C://Users/anurag/Documents/Python/Sentiment Analysis/Mask Detection/images'
```

```python
resized_shape = (224, 224)
pixelated_images = []
for index, row in df.iterrows():
    image_name = row['filename']
    img_path = os.path.join(images, image_name)
    img = Image.open(img_path)
    img = img.resize(resized_shape)
    pixels = asarray(img).astype('float32')
    pixels /= 255.0
    pixelated_images.append(pixels)
```

```python
for p in pixelated_images:
    print(np.shape(p))
```

```python
pixelated_images_refined = []
indexes = []

for index, pixels in enumerate(pixelated_images):
    if (pixels.shape == (224, 224, 4)):
        pixelated_images_refined.append(pixels)
    
    else:
        indexes.append(index)
```

```python
print(indexes)
```

```python
df = df.drop(index=indexes)
```

```python
le = LabelEncoder()

df['mask_status'] = le.fit_transform(df['mask_status'])
```

```python
X = pixelated_images_refined
y = df['mask_status']

xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3)
```

```python
xtrain = np.array(xtrain)
xtest = np.array(xtest)
ytrain = np.array(ytrain)
ytest = np.array(ytest)
```

```python
print(ytest)
```

```python
print(np.shape(xtrain))
print(np.shape(ytrain))
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Create a sequential model
model = keras.Sequential()

# Add a convolutional layer
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 4)))

# Add a max-pooling layer
model.add(layers.MaxPooling2D((2, 2)))

# Add another convolutional layer
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# Add another max-pooling layer
model.add(layers.MaxPooling2D((2, 2)))

# Flatten the output from the previous layer
model.add(layers.Flatten())

# Add a dense (fully connected) layer
model.add(layers.Dense(64, activation='relu'))

# Add an output layer with appropriate units (e.g., for classification)
model.add(layers.Dense(10, activation='softmax'))  # Change 10 to the number of classes in your dataset

# Compile the model with an appropriate optimizer, loss function, and metrics
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Print a summary of the model architecture
model.summary()

# Train the model using your data
model.fit(xtrain, ytrain, epochs=10, batch_size=32, validation_split=0.2)
```

```python
test_loss, test_acc = model.evaluate(xtest,  ytest, verbose=2)
```
